### Column-wise and multiple function application

## Group-wise operations and transformations

```python
smoker_gb = tips.groupby('smoker')['total_bill'].agg([np.mean,'std'])
smoker_gb
```

```python
tips.merge(smoker_gb, right_index=True, left_on='smoker')
```

```python
tips
```

### Apply: General split-apply-combine

```python
tips['tips_pct'] = tips['tip']/tips['total_bill']
tips
```

```python
def top(df, n=5, col='tips_pct'):
    return df.sort_values(by=col, ascending=False).head(n)
```

```python
tips_by_smoker = tips.groupby('smoker')['tips_pct'].agg([np.mean, np.std])
tips_by_smoker 
```

```python
merged = tips.merge(tips_by_smoker, left_on='smoker', right_index=True)
```

```python
top(merged)
```

```python
merged.groupby('sex').apply(top)
```

#### Suppressing the group keys

```python
tips.groupby(['sex','smoker']).apply(top)
```

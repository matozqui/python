```python
def clean_text(s):
    import re
    from unicodedata import normalize
    # -> NFD y eliminar diacríticos
    s = re.sub(
        r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", 
        normalize( "NFD", s), 0, re.I
    )
    # -> NFC
    return normalize( 'NFC', s)
```

```python
data = pd.DataFrame()
filename = './01_ ACCIDENTES POR TIPO EN  DISTRITOS.xls'
for year in range(len(pd.ExcelFile(filename).sheet_names)):
    current_sheet = pd.ExcelFile(filename).sheet_names[year]
    current_data = pd.read_excel(filename,sheet_name=current_sheet, skiprows=7, index_col=0,
             usecols="A:K", skipfooter=1)
    current_data['total_incidents'] = current_data.sum(axis=1)
    current_data['year'] = current_sheet
    data = data.append(current_data)

data.columns = data.columns.str.replace("[^\w-]","").str.lower()
data
```
